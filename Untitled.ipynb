```python
import os
import pickle
import tifffile
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr
from data_loader import get_identifier, load_all_pairs, get_all_files, load_image_pair
from segment_support import generate_weight, generate_fluorescence_labels, quantize_fluorescence
from data_assembly import generate_continuous_labels
```

```python
raw_root = 'VALIDATION/additional_lines_for_seg/line_additional-combined-24/ex0-pre/220130_213610_az24well_multiline_segmentation_D14_phase/'
pairs = load_all_pairs(raw_root)
fl_file_mapping = {get_identifier(p[0]): p for p in pairs if not p[1] is None}
```

```python
seg_root = 'VALIDATION/additional_lines_for_seg/line_additional-combined-24/ex0-pre/ex-valid-pred-to-0'
fs = sorted([f for f in os.listdir(seg_root) if f.startswith('seg_')])
segs = {}
for f in fs:
    dat = pickle.load(open(os.path.join(seg_root, f), 'rb'))
    for name, pred in zip(dat['pred_names'], dat['seg_preds']):
        _id = get_identifier(name)
        if _id in fl_file_mapping:
            segs[_id] = pred
```

```python
list(segs.keys()).index(('line_additional-combined-24', 'ex0-pre', '14', 'C2', '10'))
```

```python
k = list(segs.keys())[22]
pc = tifffile.imread(fl_file_mapping[k][0])
fl = tifffile.imread(fl_file_mapping[k][1])
```

```python
pair_dat = load_image_pair(fl_file_mapping[k])
y = quantize_fluorescence(pair_dat, np.ones_like(pair_dat[0]).astype(float), neg_percentile=10, pos_percentile=50)
_y = y[..., 1] + y[..., 2] * 2 + y[..., 3] * 3
_y = cv2.resize(_y, (384, 288))
```

```python
i = 11
__y = cv2.medianBlur((_y/3*255).astype(int).astype('uint8'), i).astype(float) / 255 * 3
plt.imshow(__y)
```
